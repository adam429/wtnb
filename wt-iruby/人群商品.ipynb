```ruby
## 引入人群  & 商品

group = Store["group_marriage"]
filter_product = Store["product_shortlist"]
nil
```

```ruby
item_删除 = group["已育"].orders.product.item.where(remove_flag:"\x01")
item_购买 = group["已育"].orders.product.item.keep.sold
item_未购买 = group["已育"].orders.product.item.keep.unsold
item_发出 = group["已育"].orders.product.item.keep

nil
```

```ruby
item_购买.size.to_f / item_发出.size
```

```ruby
puts item_删除.size, item_购买.size, item_未购买.size ,item_发出.size
```

```ruby
data_删除 = item_删除.includes(:product_spu).group(:spu).count.sort do |x,y| -1*(x[1]<=>y[1]) end.to_h
data_购买 = item_购买.includes(:product_spu).group(:spu).count.sort do |x,y| -1*(x[1]<=>y[1]) end.to_h
data_未购买 = item_未购买.includes(:product_spu).group(:spu).count.sort do |x,y| -1*(x[1]<=>y[1]) end.to_h


nil
```

```ruby
# 按照发出排序 Top 30
data = item_发出.includes(:product_spu).group(:spu).count.sort do |x,y| -1*(x[1]<=>y[1]) end.map do |x|
    [x[0],
        [x[1], 
         data_购买[x[0]] || 0,
         data_未购买[x[0]] || 0,
         data_删除[x[0]] || 0,
         (data_购买[x[0]] || 0) / ((data_购买[x[0]] || 0) + (data_未购买[x[0]] || 0)).to_f
    ]]
end.filter do |x| x[1][4]>0.2 end[0,30]

chart_data = [
    { :name=>"发出", :data=> data.map do |x| [x[0],x[1][0]] end },
]

puts_header "按照发出排序 Top 30"
puts_display column_chart(chart_data,colors:color(),stacked:true)


chart_data = [
    { :name=>"购买", :data=> data.map do |x| [x[0],x[1][1]] end },
    { :name=>"未购买", :data=> data.map do |x| [x[0],x[1][2]] end },
    { :name=>"删除", :data=> data.map do |x| [x[0],x[1][3]] end },
]
puts_display column_chart(chart_data,colors:color(),stacked:true)

Store["product_top_send"] = ProductSpu.where(spu:data.map {|x| x[0]})
nil
```
